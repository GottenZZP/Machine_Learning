## 序列化

```python
def serialize(a, b):
    return np.append(a.flatten(), b.flatten())
```

```python
theta_ser = serialize(theta1, theta2)
theta_ser.shape
```

## 解序列化

```python
def deserialize(theta_ser):
    theta1 = theta_ser[:25 * 401].reshape(25, 401)
    theta2 = theta_ser[25 * 401:].reshape(10, 26)
    return theta1, theta2
```

```python
theta1, theta2 = deserialize(theta_ser)
theta1.shape, theta2.shape
```

## 激励函数

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

## 前向传播

```python
def forward_spreed(theta_ser, X):
    theta1, theta2 = deserialize(theta_ser)
    a1 = X
    z2 = a1 @ theta1.T
    a2 = sigmoid(z2)
    a2 = np.insert(a2, 0, values=1, axis=1)
    z3 = a2 @ theta2.T
    a3 = sigmoid(z3)
    return a1, z2, a2, z3, a3
```

## 代价函数

```python
def cost(theta_ser, X, y):
    a1, z2, a2, z3, a3 = forward_spreed(theta_ser, X)
    J = -np.sum(y * np.log(a3) + (1 - y) * np.log(1 - a3)) / len(X)
    return J
```

```python
cost(theta_ser, X, y)
```

```python
def reg_cost(theta_ser, X, y, lamda):
    theta1, theta2 = deserialize(theta_ser)
    sum1 = np.sum(np.power(theta1[:, 1:], 2))
    sum2 = np.sum(np.power(theta2[:, 1:], 2))
    reg = (sum1 + sum2) * lamda / (2 * len(X))
    return cost(theta_ser, X, y) + reg
```

```python
lamda = 0.1
reg_cost(theta_ser, X, y, lamda)
```

## g(z)的求导数

```python
def sigmoid_gradient(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

## 不带正则化的梯度

```python
def gradient(theta_ser, X, y):
    theta1, theta2 = deserialize(theta_ser)
    a1, z2, a2, z3, a3 = forward_spreed(theta_ser, X)
    d3 = a3 - y
    d2 = d3 @ theta2[:, 1:] * sigmoid_gradient(z2)
    grad2 = (d3.T @ a2) / len(X)
    grad1 = (d2.T @ a1) / len(X)
    return serialize(grad1, grad2)
```

## 带正则化的梯度

```python
def reg_gradient(theta_ser, X, y, lamda):
    grad = gradient(theta_ser, X, y)
    grad1, grad2 = deserialize(grad)
    
    theta1, theta2 = deserialize(theta_ser)
    
    grad1[:, 1:] = grad1[:, 1:] + lamda * theta1[:, 1:]
    grad2[:, 1:] = grad2[:, 1:] + lamda * theta2[:, 1:]
    
    return serialize(grad1, grad2)
```

```python
def nn_training(X, y):
    initTheta = np.random.uniform(0.1, 0.5, 10285)
    result = minimize(fun=reg_cost, x0=initTheta, args=(X, y, lamda), method='TNC', jac=reg_gradient, options={'maxiter':300})
    
    return result
```

```python
lamda = 0.001
result = nn_training(X, y)
```

## 可视化隐藏层

```python
def plot_hidden(theta_ser):
    theta1,_ = deserialize(theta_ser)
    hidden = theta1[:, 1:]
    
    fig, ax = plt.subplots(figsize=(10, 10), ncols=5, nrows=5, sharex=True, sharey=True)
    
    for r in range(5):
        for c in range(5):
            ax[r, c].imshow(hidden[r * 5 + c, :].reshape(20, 20).T, cmap='gray_r')
    
    plt.xticks([])
    plt.yticks([])
```
