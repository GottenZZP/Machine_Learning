```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import random
```

```python
def initDataSet():
    dataset = load_iris()
    X = dataset.data
    y = dataset.target
    dataset = pd.DataFrame(X)
    dataset["y"] = y
    return dataset
```

```python
dataset = initDataSet()
```

```python
def randSplit(dataset, rate):
    n = dataset.shape[0]
    m = int(n * rate)
    l = list(dataset.index)
    random.shuffle(l)
    dataset.index = l
    train = dataset.loc[range(m), :]
    test = dataset.loc[range(m, n), :]
    dataset.index = range(dataset.shape[0])
    test.index = range(test.shape[0])
    return train, test
```
